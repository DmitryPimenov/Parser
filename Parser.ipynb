```python
num_of_page = 900
URL_TEMPLATE = "https://www.citilink.ru/catalog/materinskie-platy/?p="
conn = sqlite3.connect("Platy.db")

def parse(url = URL_TEMPLATE):
    result_list = {'href': [], 'title': [], 'about': []}
    for i in range(1, num_of_page + 1):
        r = requests.get(url+str(i))
        soup = bs(r.text, "html.parser")
        comp_names = soup.find_all('div', class_='ProductCardHorizontal__header-block')
        comp_info = soup.find_all('div', class_='ProductCardHorizontal__header-block')
        for name in comp_names:
            result_list['href'].append('https://www.citilink.ru'+name.a['href'])
            result_list['title'].append(name.a['title'])
        for info in comp_info:
            result_list['about'].append(info.text)
    return result_list


df = pd.DataFrame(data=parse())
df.to_sql('Platy', con=conn)
```
